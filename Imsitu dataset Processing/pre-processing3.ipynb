## Loading the whole dictionary of verbs and making the dictionary with 50 verbs

```python
dictionary_verbs = np.load('./dictionary_verbs.npy')
```

```python
np.random.shuffle(dictionary_verbs)

dictionary_fifty_verbs = dictionary_verbs[:50]
dictionary_fifty_verbs = np.sort(dictionary_fifty_verbs)
np.save('./data matrices for fifty verbs/dictionary_fifty_verbs.npy',dictionary_fifty_verbs)
```

```python
Y_fifty_verbs = []
X_name = []
for i in tqdm(os.listdir(path2images)):
    if i[-3:] == 'jpg':
        for j in range(len(dictionary_fifty_verbs)):
            if dictionary_fifty_verbs[j] in i:
                Y_fifty_verbs.append(j)
                X_name.append(i)
```
