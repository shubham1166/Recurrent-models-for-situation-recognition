## Code to make dictionary of verbs

```python
names_without_jpg = []
for i in tqdm(os.listdir(path2images)):
    if i[-3:] == 'jpg':
        ind = i.find('_')
        name = i[:ind]
        names_without_jpg.append(name)
        
dictionary_verbs = np.unique(names_without_jpg)
# np.save('./dictionary_verbs.npy',dictionary_verbs)
```

```python
dictionary_verbs[:10]
```

```python
Y_verbs = []
for i in tqdm(os.listdir(path2images)):
    if i[-3:] == 'jpg':
        for j in range(len(dictionary_verbs)):
            if dictionary_verbs[j] in i:
                Y_verbs.append(j)
```
